Задание 1

Постройте линейную диаграмму, отображающую температуру в городе за одну неделю. Создайте два списка чисел, один для оси X, представляющий дни недели, а другой для оси Y, представляющий температуру. Добавьте подписи к осям и заголовок графика.

```python
plt.plot(['M', 'T', 'W', 'Th', 'F', 'S', 'Sun'],[-2, 5, 3, -5, -4, 2, 0], c='orange', marker = 'o')
plt.title('Температура Алматы')
plt.xlabel('Дни недели')
plt.ylabel('Температура °С')
plt.legend(['1st week'], loc="upper right")
plt.grid(True)
plt.show()
```

Задание 2

Постройте точечную диаграмму, иллюстрирующую взаимосвязь между количеством часов учёбы и оценками студентов. Создайте два списка чисел, один для оси X, представляющий часы учебы, а другой для оси Y, представляющий оценки студентов. Добавьте подписи к осям и заголовок.

```python
plt.scatter([2.5, 4, 5.5, 6, 7, 8, 8.5, 10], ['C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'], c='r')
plt.title("Взаимосвязь между кол-вом часов учёбы и оценками студентов")
plt.xlabel('Часы учебы')
plt.ylabel('Оценки студентов')
plt.legend(['correlation'])
plt.grid(True, ls=':')
plt.show()
```

Задание 3

Создайте список категорий (например, разные виды спорта) и список значений (например, количество участников в каждом виде спорта). Постройте столбчатую диаграмму для сравнения количества участников. Добавьте подписи к осям и заголовок.

```python
plt.bar(['Football', 'Judo', 'Athletics', 'Tennis', 'Golf', 'Chess', 'Hockey'],[5462, 922, 1565, 2870, 408, 234, 3686], color='maroon')
plt.title('Cравнения количества участников')
plt.xlabel('Разные виды спорта')
plt.ylabel('Кол-во участников в каждом виде спорта')
plt.legend(['distribution'], loc="upper right")
plt.grid(True, ls='-.', lw = 0.5, alpha = 0.2)
plt.show()
```

Задание 4

Создайте гистограмму для анализа частоты встречаемости различных оценок. Создайте набор данных, используя генерацию случайных чисел. Например, оценки могут быть в диапазоне от 2 до 5 (включительно). Для этого используйте библиотеку numpy (функцию numpy.random.randint) Выберите подходящее количество бинов для гистограммы. Например, для оценок может подойти 8 бина. Добавьте подписи к осям и заголовок.

```python
import numpy as np
x=np.random.randint(2,6, size=8)
plt.hist(x, bins=8, color='skyblue', edgecolor='black')
plt.title('Анализ частоты оценок')
plt.xlabel('Частота оценок')
plt.ylabel('Оценки')
plt.legend(['frequency'])
plt.show()
```

Создайте две Столбчатые Диаграммы на одном изображении:

- Первая Диаграмма: Постройте столбчатую диаграмму, показывающую количество чашек разных видов кофе (например, эспрессо, латте, капучино), проданных в кофейне за день.
- Вторая Диаграмма: Создайте столбчатую диаграмму для пяти различных закусок (например, круассаны, багеты, печенья и т.д.), также проданных в этот же день

```python
y=np.random.randint(50, size=8)
y1=np.random.randint(15, size=5)
```

```python
plt.figure(figsize=(17, 6))

plt.subplot(1, 2, 1)
plt.bar(['Espresso', 'Latte', 'Cappuccino', 'Americano', 'Mocha', 'Macchiato', 'Flat White', 'Affogato'], y, color='darkgreen')
plt.title('Количество чашек разных видов кофе')

plt.subplot(1, 2, 2)
plt.bar(['Круассаны', 'Багеты', 'Брускетты', 'Печенья', 'Тосты'], y1, color='royalblue')
plt.title('Различные Закуски')

plt.show()
```
