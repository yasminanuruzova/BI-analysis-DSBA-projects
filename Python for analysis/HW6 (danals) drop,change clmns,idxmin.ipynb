2. Удалите колонку 'LikePresentations', она нам не понадобится

```python
df.drop('LikePresentation', axis=1, inplace=True)
df.head()
```

3. Измените названия ваших колонок - разделите слова в названии колонок. Например, замените AcademicPerformance на Academic Performance



```python
columns=list(df.columns)
```

```python
columns[2:10]= ['Academic Performance','Taking Note In Class','Depression Status','Face Challanges To Complete Academic Task','Sleep Per Day Hours','Number Of Friend','Like New Things']
df.columns=columns
df.head()
```

4.Очистка данных. Удалите дубликаты, если они есть и строки с пропущенными значениями



```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
df=df.dropna()
df.head()
```

```python
df.shape
```

5.Выявить среднее количество сна студентов

```python
df.groupby('Gender')[ 'Sleep Per Day Hours'].mean().reset_index()
```

6.Вывести таблицу, где только студенты с оценками "excellent" и посчитать среднее значение сна

```python
df[df['Academic Performance'] == 'Excellent'].groupby('Academic Performance').agg({'Sleep Per Day Hours': 'mean'})
```

```python
#2й способ
df[df['Academic Performance'] == 'Excellent'].groupby('Academic Performance')['Sleep Per Day Hours'].mean().to_frame()
```

```python
df[df['Academic Performance'] == 'Excellent']['Sleep Per Day Hours'].mean()
```

7. Вывести таблицу, где только студенты с депрессией - yes и посчитать среднее значение их сна

```python
depr=df['Depression Status']=='Yes'
```

```python
df[depr].groupby('Depression Status').agg({'Sleep Per Day Hours':'mean'})
```
