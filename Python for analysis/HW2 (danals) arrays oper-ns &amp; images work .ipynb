```python
#4
import matplotlib.pyplot as plt
from skimage import data, color

image = data.cat()


blue_white_image = np.copy(image)
blue_white_image[:, :, 0] = 50 
blue_white_image[:, :, 1] = 20

bright_areas = blue_white_image[:, :, 2] > 500
blue_white_image[bright_areas, 2] = 255

# Display the original and the modified image
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blue_white_image)
plt.title('Blue and White Image')
plt.axis('off')

plt.show()
```
